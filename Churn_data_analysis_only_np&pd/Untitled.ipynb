```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Sample Data
df = pd.read_csv('churn_modelling.csv')  # Replace with actual path

# Initialize Dash app
app = dash.Dash(__name__)
app.title = "Bank Customer Dashboard"

# Layout
app.layout = html.Div([
    html.H1("Customer Churn Dashboard", style={'textAlign': 'center'}),

    # Dropdown for filtering by Geography
    html.Div([
        html.Label("Select Country:"),
        dcc.Dropdown(
            id='geo-dropdown',
            options=[{'label': geo, 'value': geo} for geo in df['Geography'].unique()],
            value='France',
            clearable=False
        )
    ], style={'width': '30%', 'margin': 'auto'}),

    # Graphs
    html.Div([
        dcc.Graph(id='churn-by-geo'),
        dcc.Graph(id='credit-by-gender'),
        dcc.Graph(id='balance-pie'),
        dcc.Graph(id='age-dist')
    ])
])

# Callbacks
@app.callback(
    Output('churn-by-geo', 'figure'),
    Output('credit-by-gender', 'figure'),
    Output('balance-pie', 'figure'),
    Output('age-dist', 'figure'),
    Input('geo-dropdown', 'value')
)
def update_graphs(selected_geo):
    churn_fig = px.bar(
        df.groupby('Geography')['Exited'].mean().reset_index(),
        x='Geography', y='Exited',
        title='Churn Rate by Geography'
    )

    credit_fig = px.bar(
        df.groupby('Gender')['CreditScore'].mean().reset_index(),
        x='Gender', y='CreditScore',
        title='Average Credit Score by Gender',
        color='Gender'
    )

    balance_fig = px.pie(
        df, names='Exited', values='Balance',
        title='Balance Distribution by Churn Status',
        hole=0.4,
        labels={0: 'Stayed', 1: 'Exited'}
    )

    age_fig = px.histogram(
        df[df['Geography'] == selected_geo],
        x='Age',
        nbins=20,
        title=f'Age Distribution in {selected_geo}'
    )

    return churn_fig, credit_fig, balance_fig, age_fig

# Run app
if __name__ == '__main__':
    app.run_server(debug=True)
```
